```python
import numpy as np

def cal_csi(Obs, Pre, Tre=0):
    # Tre: rain/no rain threshold
    # POD(Probability of Detection),FOH(frequency of hit)
    # FAR(False Alarm Ratio), CSI(Critical Success Index)
    # HSS(Heidke skillscore),Ebert et al. [2007]
    if len(Obs) > 1:
        n11 = np.sum((Obs > Tre) & (Pre > Tre))
        n10 = np.sum((Obs <= Tre) & (Pre > Tre))
        n01 = np.sum((Obs > Tre) & (Pre <= Tre))
        # n00 = np.sum((Obs <= Tre) & (Pre <= Tre))
    try:
        CSI = n11 / (n11 + n01 + n10)
    except:
        CSI = np.nan
    return CSI

def cal_bs(Obs, Pre):
    return np.nanmean((Obs-Pre)**2)

gmet_stndatafile = '/Users/localuser/Research/EMDNA/stndata_whole.npz'
path_pop = '/Users/localuser/Research/EMDNA/pop'
file_reapop_stn = path_pop + '/reanalysis_pop_stn.npz'

datatemp = np.load(gmet_stndatafile)
stndata = datatemp['prcp_stn']
stnlle = datatemp['stn_lle']
nstn, ntimes = np.shape(stndata)
del datatemp
stndata[stndata>0]=1

datatemp = np.load(file_reapop_stn)
reapop_stn = datatemp['reapop_stn']
del datatemp
```

```python
csi=np.nan * np.zeros([nstn,3])
bs=np.nan * np.zeros([nstn,3])
for i in range(3):
    print(i)
    for j in range(nstn):
        if np.isnan(stndata[j,0]):
            continue
        csi[j,i]=cal_csi(stndata[j,:],reapop_stn[i,j,:],0.5)
        bs[j,i]=cal_bs(stndata[j,:],reapop_stn[i,j,:])
```

```python
from matplotlib import pyplot as plt
dplot = csi
titles=['CSI: ERA5','CSI: MERRA2','CSI: JRA55']
plt.figure(figsize=[14,7])
for i in range(len(titles)):
    plt.subplot(2,3,i+1)
    plt.scatter(stnlle[:,1],stnlle[:,0],5,dplot[:,i],cmap='jet')
    cbar = plt.colorbar()
    plt.clim([0,1])
    plt.title(titles[i])

dplot = bs
titles=['BS: ERA5','BS: MERRA2','BS: JRA55']
for i in range(len(titles)):
    plt.subplot(2,3,i+4)
    plt.scatter(stnlle[:,1],stnlle[:,0],5,dplot[:,i],cmap='jet')
    cbar = plt.colorbar()
    plt.clim([0,0.3])
    plt.title(titles[i])
```
