```python
%%time

import numpy as np
import time
from multiprocessing import Pool, RawArray

# A global dictionary storing the variables passed from the initializer.
var_dict = {}

def init_worker(d1, d2):
    # Using a dictionary is not strictly necessary. You can also
    # use global variables.
    var_dict['d1'] = d1
    var_dict['d2'] = d2

def worker_func(i):
    time.sleep(1)
    return np.sum(var_dict['d1'][:,i]*var_dict['d2'][:,i])

# We need this check for Windows to prevent infinitely spawning new child
# processes.
if __name__ == '__main__':
    dd1 = np.random.rand(3, 10)
    dd2 = np.random.rand(3, 10)
    with Pool(processes=5, initializer=init_worker, initargs=(dd1, dd2)) as pool:
        result = pool.map(worker_func, range(10))
        print(result)
```
