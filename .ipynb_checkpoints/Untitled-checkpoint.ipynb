```python
import numpy as np
import auxiliary as au
import pandas
from matplotlib import pyplot as plt
import tabulate

def cal_statiscs(file_stn, file_oimerge, var):
    # load station data
    d1=np.load(file_stn)
    stndata=d1[var+'_stn']
    stndata[np.nanmean(stndata,axis=1)==0,:]=np.nan
    nstn,ntimes=np.shape(stndata)
    d2=np.load(file_oimerge)
    oidata=d2['oimerge_stn']
    # calculate the metrics for each station
    met_prcp=np.nan * np.zeros([nstn,4])
    for i in range(nstn):
        if np.isnan(stndata[i,0]):
            continue
        met_prcp[i,:] = au.metric(stndata[i],oidata[i,:])
    return met_prcp

file_stn='/datastore/GLOBALWATER/CommonData/EMDNA_new/stndata_aftercheck.npz'
file_regression='/datastore/GLOBALWATER/CommonData/EMDNA_new/stn_reg_aftercheck/regression_stn.npz'

file_oimerge='/datastore/GLOBALWATER/CommonData/EMDNA_new/OImerge_GWRQMBMA/OImerge_stn_GWRQMBMA_prcp.npz'
met_prcp1 = cal_statiscs(file_stn, file_oimerge, 'prcp')
file_oimerge='/datastore/GLOBALWATER/CommonData/EMDNA_new/OImerge_GWRLSBMA/OImerge_stn_GWRLSBMA_prcp.npz'
met_prcp2 = cal_statiscs(file_stn, file_oimerge, 'prcp')

file_oimerge='/datastore/GLOBALWATER/CommonData/EMDNA_new/OImerge_GWRQMBMA/OImerge_stn_GWRQMBMA_tmean.npz'
met_tmean1 = cal_statiscs(file_stn, file_oimerge, 'tmean')
file_oimerge='/datastore/GLOBALWATER/CommonData/EMDNA_new/OImerge_GWRLSBMA/OImerge_stn_GWRLSBMA_tmean.npz'
met_tmean2 = cal_statiscs(file_stn, file_oimerge, 'tmean')
file_oimerge='/datastore/GLOBALWATER/CommonData/EMDNA_new/OImerge_GWRQMBMA/OImerge_stn_GWRQMBMA_trange.npz'
met_trange1 = cal_statiscs(file_stn, file_oimerge, 'trange')
file_oimerge='/datastore/GLOBALWATER/CommonData/EMDNA_new/OImerge_GWRLSBMA/OImerge_stn_GWRLSBMA_trange.npz'
met_trange2 = cal_statiscs(file_stn, file_oimerge, 'trange')

np.savez_compressed('OImetric.npz',met_prcp_QM=met_prcp1,met_prcp_LS=met_prcp2, met_tmean_QM=met_tmean1,met_tmean_LS=met_tmean2,met_trange_QM=met_trange1,met_trange_LS=met_trange2)
```

```python
import numpy as np
import netCDF4 as nc
file='/Users/localuser/Research/PBCOR_V1.0/CHPclim_V1.nc'
datatemp = nc.Dataset(file)
pclim = datatemp['corr_P_monthly'][:].data
latclim = datatemp['lat'][:].data
lonclim = datatemp['lon'][:].data
```

```python
from matplotlib import pyplot as plt
plt.imshow(pclim[0])
plt.clim([0,100])
```

```python
FileGridInfo = '/Users/localuser/Research/EMDNA/basicinfo/gridinfo_whole.nc'
ncfid = nc.Dataset(FileGridInfo)
gridlat = ncfid.variables['latitude'][:].data
gridlon = ncfid.variables['longitude'][:].data
gridele = ncfid.variables['elev'][:].data
gridgns = ncfid.variables['gradient_n_s'][:].data
gridgwe = ncfid.variables['gradient_w_e'][:].data
mask = ncfid.variables['mask'][:].data  # 1: grids to be considered; the other values: invalid grids
ncfid.close()

gridinfo = np.zeros([nrows, ncols, 6])
gridinfo[:, :, 0] = 1
gridinfo[:, :, 1] = gridlat
gridinfo[:, :, 2] = gridlon
gridinfo[:, :, 3] = gridele
gridinfo[:, :, 4] = gridgns
gridinfo[:, :, 5] = gridgwe
del gridlat, gridlon, gridele, gridgns, gridgwe
```

```python
pclim_tar = np.nan * np.zeros([nrows, ncols, 12])
pclim_num = np.nan * np.zeros([nrows, ncols])
res = 0.1/2 # half resolution
for rr in range(nrows):
    if np.mod(rr,50)==0:
        print(rr, nrows)
    for cc in range(ncols):
        if mask[rr, cc] != 1:
            continue
        latrc = gridinfo[rr, cc, 1]
        lonrc = gridinfo[rr, cc, 2]
        lonind = np.where((lonclim >= lonrc-res) & (lonclim <= lonrc+res))[0]
        latind = np.where((latclim >= latrc - res) & (latclim <= latrc + res))[0]
        pclim_rc = pclim[:, latind[0]:latind[1]+1, lonind[0]:lonind[1]+1]
        pclim_num[rr, cc] = np.sum(~np.isnan(pclim_rc[0, :, :]))
        for m in range(12):
            pclim_tar[rr, cc, m] = np.nanmean(pclim_rc[m, :, :])
```
